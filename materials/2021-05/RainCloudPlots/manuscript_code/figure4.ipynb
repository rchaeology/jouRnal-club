### Figure 4 - simulating data where a raincloud plot can reveal hidden distributional information. 

```python
import matplotlib.pylab as plt
%matplotlib inline
import numpy as np
import os
import pandas as pd
import ptitprince as pt
import seaborn as sns
sns.set(style="whitegrid",font_scale=2)
```

```python
x = (3.5+2*np.random.randn(1000, 2)); x1 = (-3.5+2*np.random.randn(1000, 2))
x=np.concatenate((x,x1),axis=0)
x_cat = np.random.randint(0,4,size=(2000, 1))
x=np.concatenate((x_cat,x),axis=1)
df_rand = pd.DataFrame(x, columns=['Category','A','B'])
```

```python
# Make the figures directory if it doesn't yet exist
if not os.path.isdir('../figs/'):
    os.makedirs('../figs/')
```

```python
fig, ax = plt.subplots(figsize=(12,11))

ax = pt.RainCloud(data = df_rand, x = "Category", y = "A",orient='h',bw=.1, ax=ax)
sns.despine()

fig.savefig('../figs/figure_4.jpg')
```
